```python
import pandas as pd
import numpy as np
import os
import datetime
import sklearn.metrics
import sklearn as sk
```

```python
positive_crime_data=pd.read_csv('/haizhi/dataset/CNT_JiangHan_1000.csv')
negative_crime_data=pd.read_csv('/haizhi/dataset/CNT_JiangHan_Zero_1000.csv')
```

```python
positive_crime_data['lon']=positive_crime_data['lon']/1000
positive_crime_data['lat']=positive_crime_data['lat']/1000
```

```python
negative_crime_data=negative_crime_data.rename(columns={'lon_zero':'lon','lat_zero':'lat'})
negative_crime_data['lon']=negative_crime_data['lon']/1000
negative_crime_data['lat']=negative_crime_data['lat']/1000
```

```python
beginning=datetime.datetime.strptime('2016-01-01 0','%Y-%m-%d %H')
```

```python
positive_crime_data['delta_days']=positive_crime_data['hours']//24
positive_crime_data['hour']=positive_crime_data['hours']%24
negative_crime_data['delta_days']=negative_crime_data['hours']//24
negative_crime_data['hour']=negative_crime_data['hours']%24
```

```python
def hours2datetime(line):
    beginning=datetime.datetime.strptime('2016-01-01 0','%Y-%m-%d %H')
    #print(line['hours'])
    delta=datetime.timedelta(line['delta_days'])
    current=beginning+delta
    line['date']=str(current.date())
    return line
    #print(current.date(),current.hour)
```

```python
negative_crime_data_date=negative_crime_data.apply(hours2datetime,axis=1)
positive_crime_data_date=positive_crime_data.apply(hours2datetime,axis=1)
```

```python
concated_data=pd.concat([positive_crime_data_date,negative_crime_data_date],ignore_index=True)
concated_data=concated_data.drop(columns=['delta_days','hours'])
concated_data['date']=concated_data['date'].apply(lambda x:str(x))
weather_data=pd.read_csv('/root/HY/wuhan_2016_2018.csv')
weather_data=weather_data.drop(columns=['Unnamed: 0','city','province'])
merged_weather=pd.merge(concated_data,weather_data,how='left')
date_data=pd.read_csv('/haizhi/dataset/crime_date_preprocess.csv')
```

```python
merged_weather_date=pd.merge(merged_weather,date_data,how='left')
merged_weather_date=merged_weather_date.drop(columns=['date','wd','wf'])
```

```python
import numpy as np
#merged_array=np.array(merged_weather_date)
wearther_list=[u'中雨', u'中雨~多云', u'中雨~大雨', u'中雨~小雨', u'中雨~晴', u'中雨~阴', u'中雨~阵雨',
       u'中雪~小雪', u'多云', u'多云~中雨', u'多云~小雨', u'多云~晴', u'多云~阴', u'多云~阵雨',
       u'多云~雷阵雨', u'大雨', u'大雨~中雨', u'大雨~小雨', u'大雨~暴雨', u'大雨~阴', u'大雪~中雪',
       u'小雨', u'小雨~中雨', u'小雨~冻雨', u'小雨~多云', u'小雨~大雨', u'小雨~晴', u'小雨~阴',
       u'小雨~阵雨', u'小雨~雨夹雪', u'小雪~中雪', u'小雪~多云', u'小雪~大雪', u'小雪~阴', u'晴',
       u'晴~多云', u'晴~小雨', u'晴~阴', u'暴雨~大雨', u'阴', u'阴~中雨', u'阴~多云', u'阴~小雨',
       u'阴~晴', u'阴~雨夹雪', u'阵雨', u'阵雨~中雨', u'阵雨~多云', u'阵雨~大雨', u'阵雨~小雨',
       u'阵雨~阴', u'雨夹雪~中雪', u'雨夹雪~多云', u'雨夹雪~小雪', u'雨夹雪~晴']
i=0
weather2number={}
for j in wearther_list:
    weather2number[j]=i
    i+=1
merged_weather_date['weather']=merged_weather_date['weather'].apply(lambda x:weather2number[x.decode('utf-8')])
quailty_list=['严重污染','中度污染','优','良','轻度污染','重度污染']
i=0
quality2number={}
for j in quailty_list:
    quality2number[j]=i
    i+=1
merged_weather_date['quality']=merged_weather_date['quality'].apply(lambda x:quality2number[x])
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn import metrics
```

```python
import matplotlib.pyplot as plt
def plot_roc_auc(model, y_test_true, x_test):
    x_ = np.linspace(0, 1, 10)
    y_test_pred_prob = model.predict_proba(x_test)[:, 1]
    y_test_pred = model.predict(x_test)
    fpr, tpr, thresholds = sk.metrics.roc_curve(y_test_true, y_test_pred_prob)
    plt.plot(x_, x_)
    plt.plot(fpr, tpr)
    plt.grid()
    plt.legend(['y=x', 'roc'])
    plt.xlabel('FPR')
    plt.ylabel('TPR')
    plt.text(0.6, 0.3, 'accuracy = {}'.format(sk.metrics.accuracy_score(y_test_true, y_test_pred)))
    plt.text(0.6, 0.2, 'auc = {}'.format(sk.metrics.auc(fpr, tpr)))
    plt.show()
```

```python
def plot_roc_auc_list(model_list, y_test_true, x_test):
    x_ = np.linspace(0, 1, 10)
    legend_list=['y=x']
    for i in range(len(model_list)):
        model=model_list[i]
        y_test_pred_prob = model.predict_proba(x_test)[:, 1]
        y_test_pred = model.predict(x_test)
        fpr, tpr, thresholds = sk.metrics.roc_curve(y_test_true, y_test_pred_prob)
        plt.plot(fpr, tpr)
        legend_list.append( 'roc'+str(i*10+10))
#         plt.text(0.6, 0.3, str(i*10+10)+' accuracy = {}'.format(sk.metrics.accuracy_score(y_test_true, y_test_pred)))
#         plt.text(0.6, 0.2, str(i*10+10)+' auc = {}'.format(sk.metrics.auc(fpr, tpr)))
        print(str(i*10+10)+' accuracy = {}'.format(sk.metrics.accuracy_score(y_test_true, y_test_pred)))
        print(str(i*10+10)+' auc = {}'.format(sk.metrics.auc(fpr, tpr)))
    #plt.legend([])
    plt.plot(x_, x_)
    plt.grid()
    plt.legend(legend_list)
    plt.xlabel('FPR')
    plt.ylabel('TPR')
    plt.show()
```

```python
def plot_roc_auc_list(model_list, y_test_true, x_test,legend_list_input):
    x_ = np.linspace(0, 1, 10)
    legend_list=['y=x']
    legend_list.extend(legend_list_input)
    for i in range(len(model_list)):
        model=model_list[i]
        y_test_pred_prob = model.predict_proba(x_test)[:, 1]
        y_test_pred = model.predict(x_test)
        fpr, tpr, thresholds = sk.metrics.roc_curve(y_test_true, y_test_pred_prob)
        plt.plot(fpr, tpr)
#         legend_list.append( 'roc'+str(i*10+10))
#         plt.text(0.6, 0.3, str(i*10+10)+' accuracy = {}'.format(sk.metrics.accuracy_score(y_test_true, y_test_pred)))
#         plt.text(0.6, 0.2, str(i*10+10)+' auc = {}'.format(sk.metrics.auc(fpr, tpr)))
        print(str(legend_list_input[i])+' accuracy = {}'.format(sk.metrics.accuracy_score(y_test_true, y_test_pred)))
        print(str(legend_list_input[i])+' auc = {}'.format(sk.metrics.auc(fpr, tpr)))
    #plt.legend([])
    plt.plot(x_, x_)
    plt.grid()
    plt.legend(legend_list)
    plt.xlabel('FPR')
    plt.ylabel('TPR')
    plt.show()
```
